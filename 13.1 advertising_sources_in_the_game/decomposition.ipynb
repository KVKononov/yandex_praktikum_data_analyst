# Анализ рекламных источников в игре

## Описание проекта

Основной датасет содержит данные о событиях, совершенных в мобильной игре «Космические братья». В ней пользователи строят свою космическую программу и пытаются преуспеть в нелёгком деле колонизации галактики. Основная монетизация игры — только планируется. Но предполагается, что в приложении будет происходить показ рекламы на экране с выбором типа объекта для постройки. В датасете представлены данные по игре пользователей на первом уровне. Завершение первого уровня требует от игрока выполнения одного из двух условий:

- Победа над первым врагом
- Реализация проекта - разработка орбитальной сборки спутников

В датасете содержатся данные первых пользователей приложения — когорты пользователей, которые начали пользоваться приложением в период с 4 по 10 мая включительно.

Информация выдана от менеджера продукта.

Задача: на основании поведения пользователей изучить рекламные источники, для дальнейшего анализа окупаемости игры.

Проверить следующие гипотезы:
1. Время завершения уровня различается в зависимости способа прохождения.
2. Количество построенных зданий различается у пользователей с разных рекламных каналов.

## Описание данных

Датасет user_source.csv:

* `user_id` - идентификатор пользователя
* `source` - источник, с которого пришёл пользователь, установивший приложение

Датасет game_actions.csv:

* `event_datetime` — время события
* `event` — одно из трёх событий:
    1. building — объект построен
    2. finished_stage_1 — первый уровень завершён
    3. project — проект завершён


* `building_type` — один из трёх типов здания:
    1. assembly_shop — сборочный цех
    2. spaceport — космопорт
    3. research_center — исследовательский центр


* `user_id` — идентификатор пользователя
* `project_type` — тип реализованного проекта

Датасет ad_cost.csv:

* `day` - день, в который был совершен клик по объявлению
* `source` - источник трафика
* `cost` - стоимость кликов

Путь к файлу: user_source.csv

Путь к файлу: game_actions.csv

Путь к файлу: ad_costs.csv

### Предобработка данных

При необходимости провести предобработку. Проверить наличие пропусков и дубликатов.

### Исследовательский анализ данных

1. Изучить стратегию пользователей. Сравнить количество игроков для обоих вариантов (победа над первым врагом и реализация проекта - разработка орбитальной сборки спутников).
2. Проверить взаимосвязь рекламного источника и стратегии.
3. Изучить количество построенных зданий у пользователей с разных рекламных каналов.
4. Проверить активность игроков с разных рекламных каналов.

### Проверка гипотез

1. Время завершения уровня различается в зависимости способа прохождения.
2. Количество построенных зданий различается у пользователей с разных рекламных каналов.
